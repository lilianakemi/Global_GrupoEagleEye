```python
import pandas as pd
import numpy as np
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Professor Hellyson/Global/dataset_tomate2.csv')
data.dropna(axis=1,inplace=True)
data.drop('id', axis = 1, inplace=True)
data.head(133)
```

```python
  import numpy as np
  from sklearn.preprocessing import MinMaxScaler
```

```python
def normaliza(coluna):

  X = np.array(data[coluna]).reshape(-1,1)
  scaler = MinMaxScaler()
  scaler.fit(X)
  X_scaled = scaler.transform(X)
  data[coluna] = X_scaled.reshape(1,-1)[0]
```

```python
normaliza('trat')
normaliza('MCARI1_d28')
normaliza('SR_d28')
normaliza('MCARI1_d01')
normaliza('MCARI1_d04')
normaliza('SR_d01')
normaliza('MCARI1_d08')
normaliza('SR_d08')
normaliza('SR_d04')
```

```python
x_uni = data[['trat', 'NDVI_d28', 'SAVI_d28', 'GNDVI_d28', 'MCARI1_d28', 'SR_d28', 'NDVI_d01', 'SAVI_d01', 'GNDVI_d01', 'MCARI1_d01', 'SR_d01', 'NDVI_d04', 'SAVI_d04', 'GNDVI_d04', 'MCARI1_d04','SR_d04','NDVI_d08', 'SAVI_d08', 'GNDVI_d08', 'MCARI1_d08', 'SR_d08']]
y_uni = data['Classe']

```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier

x_train,x_test,y_train,y_test  = train_test_split(x_uni, y_uni,
                                
                                                    test_size = 0.3, 
                                                    random_state = 2)
```

```python
print('Training data shape', x_train.shape)
print('Training labels shape', y_train.shape)
print('Test data shape', x_test.shape)
print('Test labels shape', y_test.shape)
```
